```python
c_test.drop(['V86'],axis = 1)
```

```python
from sklearn.model_selection import train_test_split
c_train1,c_train2 = train_test_split(c_train,test_size= 0.2,random_state = 2)
x_train=c_train1.drop(['V86'],1)
y_train=c_train1['V86'].astype(int)
x_test=c_train2.drop(['V86'],1)
y_test=c_train2['V86'].astype(int)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(class_weight='balanced', criterion='entropy',
                       max_features=75, min_samples_leaf=20, min_samples_split=20,
                       n_estimators=500)
```

```python
clf.fit(x_train, y_train)
```

```python
predict_rf = clf.predict(x_test).astype(int)
predict_rf
```

```python
from sklearn.metrics import fbeta_score
```

```python
fbeat2 = fbeta_score(y_test,predict_rf,beta=2,average = "binary")
```
