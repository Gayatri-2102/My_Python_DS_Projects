```python
cat_cols1 = ['Medicine_Type']
```

```python
for col in cat_cols1:
    freqs=c_all[col].value_counts()
    k=freqs.index[freqs>500][:-1]
    for cat in k:
        name=col+'_'+cat
        c_all[name]=(c_all[col]==cat).astype(int)
    del c_all[col]
    print(col)
    
```

```python
cat_cols2 = ['SidEffect_Level', 'Area_Type', 'Area_City_Type',
       'Area_dist_level']
```

```python
for col in cat_cols2:
    dummy=pd.get_dummies(c_all[col],drop_first=True,prefix=col)
    c_all=pd.concat([c_all,dummy],axis=1)
    del c_all[col]
    print(col)
del dummy
```
